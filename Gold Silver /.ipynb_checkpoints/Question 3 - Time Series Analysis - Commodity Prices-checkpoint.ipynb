# Fetching historical commodity data from investing.com

```python
# Written by Bulent Arslan

import requests
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as sf
import statsmodels.tsa.api as smt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.graphics.gofplots import qqplot
from arch import arch_model

%matplotlib inline


date_start = '06/01/2018' # Query start time, MM/DD/YYYY
date_end = '09/01/2018'   # Query end time, MM/DD/YYYY
curr_id = '8830'          # Commodity ID, 8830 for gold, 8836 for silver, 18 for USD/TRY

s = requests.session()
url = 'https://www.investing.com/instruments/HistoricalDataAjax'
headers={'Origin': 'http://www.investing.com',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.100 Safari/537.36',
        'Date':'Tue, 21 Oct 2018 03:09:55 GMT',
        'Content-Type': 'application/x-www-form-urlencoded',
        'X-Requested-With': 'XMLHttpRequest'}
data = {'curr_id': curr_id,
'st_date': date_start,
'end_date': date_end,
'interval_sec': 'Daily',
'sort_ord': 'DESC',
'action': 'historical_data'}
r = s.post(url,headers=headers,data=data)
data = pd.read_html(r.content)
df =data[0]
df.Date = pd.to_datetime(df.Date)
df = df.set_index(df.Date)
df=df.resample('b').mean()
df.Price.plot(figsize=(12,4))
plt.xlabel('Date')
plt.ylabel('USD per oz')
plt.title('Historical Gold Data')
```

# Part 1 : Are gold prices good predictors for their future prices ?


We will examine Histogram, ACF, PACF and QQ plots to get the initial insights whether historical commodity data is a good predictor or not.

```python
f, axes = plt.subplots(2,2,figsize=(16,12))
plot_acf(df.Price,lags=40, ax = axes[0,0])
plot_pacf(df.Price,lags=25, ax = axes[0,1])
qqplot(df.Price,fit=True,line='45', ax = axes[1,0])
axes[1,1] = df.Price.hist(bins=12)
axes[1,1].set_title('Histogram of gold prices')
f.suptitle('Histogram, ACF, PACF and QQ plots')
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 11, 9

decomposition = smt.seasonal_decompose(df.Price, model='additive')
fig = decomposition.plot()
plt.show()
```

It can be inferred from the plots that historical prices of the commodity are not random variables. Therefore, the data can be modeled as a function of its past values.   

There is a clear trend in data, so we start modeling by differentiating to eliminate trend.  
In order to eliminate trend, we will fit an ARIMA(0,2,0) model and observe ACF, PACF and QQ plots.

```python
diff_model = smt.ARIMA(df.Price,order=(0,2,0))
diff_result = diff_model.fit(trend='c')

f, axes = plt.subplots(2,2,figsize=(16,12))
plot_acf(diff_result.resid,lags=20, ax = axes[0,0])
plot_pacf(diff_result.resid,lags=12, ax = axes[0,1])
qqplot(diff_result.resid,fit=True,line='45', ax = axes[1,0])
axes[1,1] = diff_result.resid.hist(bins=12)
axes[1,1].set_title('Histogram of gold prices')
f.suptitle('Histogram, ACF, PACF and QQ plots')
```

Even after eliminating trend, we see that the data is still not stationary by observing plots above.  
Autocorrelation exists and this indicates past values of commodity price can be a good predictor for future values.

Here we define a grid search function to find best parameters for p,q and d:

```python
def _get_best_model(TS,pq_rng=5,d_rng=2):
    '''Finds best parameters for given series to be used in
    GARCH Model fitting. Parameters are suggested based on
    lowest AIC.
    best_aic, best_order, best_mdl = _get_best_model(TS,pq_rng,d_rng)
    ---------------
    Returns:
    best_aic: Lowest AIC obtained in grid search
    best_order: (p,d,q) values giving the lowest AIC
    best_mdl: statsmodels.tsa.api.ARIMA object with best parameters
    ---------------
    Parameters:
    TS: time series object
    pq_rng: (p,q) range to perform search within
    d_rng: d range to perform search within'''
    best_aic = np.inf
    best_order = None
    best_mdl = None

    pq_rng = range(pq_rng) # [0,1,2,3,4]
    d_rng = range(d_rng) # [0,1]
    for i in pq_rng:
        for d in d_rng:
            for j in pq_rng:
                try:
                    tmp_mdl = smt.ARIMA(TS, order=(i,d,j)).fit(
                        method='mle', trend='c'
                    )
                    tmp_aic = tmp_mdl.aic
                    if tmp_aic < best_aic:
                        best_aic = tmp_aic
                        best_order = (i, d, j)
                        best_mdl = tmp_mdl
                except: continue
    print('aic: {:6.5f} | order: {}'.format(best_aic, best_order))
    return best_aic, best_order, best_mdl
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    best_aic, best_order, best_model = \
    _, best_order, best_model = _get_best_model(df['Price'],pq_rng=10,d_rng=2)
print('best AIC:',best_aic)
print('best order:',best_order)
```

```python
p,d,q = 5,1,3
best_model = smt.ARIMA(df.Price,order=(p,d,q))
best_result = best_model.fit(method='mle',trend='c')
best_result.summary()
```

```python
residuals = best_result.resid

f, axes = plt.subplots(2,2,figsize=(16,12))
plot_acf(residuals,lags=40, ax = axes[0,0])
plot_pacf(residuals,lags=25, ax = axes[0,1])
qqplot(residuals,fit=True,line='45', ax = axes[1,0])
axes[1,1] = residuals.hist(bins=12)
axes[1,1].set_title('Histogram of gold prices')
f.suptitle('Histogram, ACF, PACF and QQ plots')
```

# Part 2 : Increase or Decrease Gold Price Prediction

## ARMA(1,1) Model - Autoregressive Moving Average

### AIC : 907

```python
df['Diff'] = (df.Price.diff()>0).astype(float)
df['y'] = df.Diff
(p,d,q) = (5,0,3)
model_ar = smt.ARIMA(df.y,order=(p,0,q))
result_ar = model_ar.fit(method='css',trend='c')
```

```python
def sigmoid(x, derivative=False):
  return x*(1-x) if derivative else 1/(1+np.exp(-x))
```

```python
resid_ar = result_ar.resid

f, axes = plt.subplots(2,2,figsize=(16,12))
plot_acf(resid_ar,lags=40, ax = axes[0,0])
plot_pacf(resid_ar,lags=25, ax = axes[0,1])
qqplot(resid_ar,fit=True,line='45', ax = axes[1,0])
axes[1,1] = resid_ar.hist(bins=12)
axes[1,1].set_title('Histogram of gold prices')
f.suptitle('Histogram, ACF, PACF and QQ plots')

y_pred = sigmoid(result_ar.predict())>0.9
result_ar.summary()
```

```python
# Accuracy Score of Increase or Decrease Gold Price Prediction

from sklearn.metrics import accuracy_score
accuracy_score(df.y.astype(bool)[:-5],y_pred)
```
